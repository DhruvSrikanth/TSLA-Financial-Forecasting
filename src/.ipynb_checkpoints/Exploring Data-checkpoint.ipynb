```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/processed_dataset/dataset.csv")
df = df.set_index('Date')
df.head()
```

```python
plt.figure(figsize=(20, 8))
df['Close'].plot()
plt.show()
```

```python
df['PriceDiff'] = df['Close'].shift(-1) - df['Close'] # Calculate price difference between closing prices by day
df['Return'] = df['PriceDiff'] / df['Close'] # Calculate return based on price difference
df['MA10'] = df['Close'].rolling(10).mean() # Calculate moving average based on a 10 day window
df['MA50'] = df['Close'].rolling(50).mean() # Calculate moving average based on a 50 day window
df['Shares'] = [1 if df.loc[ei, 'MA10'] > df.loc[ei, 'MA50'] else 0 for ei in df.index] # Calculate shares
df['Close1'] = df['Close'].shift(-1) # Close price of tomorrow
df['Profit'] = [df.loc[ei, 'Close1'] - df.loc[ei, 'Close'] if df.loc[ei, 'Shares'] == 1 else 0 for ei in df.index] # Calculate profit gained if the shares are 1, otherwise 0
df['wealth'] = df['Profit'].cumsum() # Amassed wealth by hedging our bets on stock price rising
```
