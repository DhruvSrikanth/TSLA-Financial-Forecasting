### Process Stocks Dataset

```python
import pandas as pd
```

```python
tsla_df = pd.read_csv('data/raw/TSLA_stock/TSLA.csv')
tsla_df.head()
```

```python
def restrict_data(df, end = '2020'):
    dates = df.Date.tolist()
    for i in range(len(dates)):
        if end in dates[i]:
            i -= 1
            break
    df = df.iloc[:i+1]
    df.reset_index()
    return df
```

```python
restricted_tsla_df = restrict_data(tsla_df, '2020')
restricted_tsla_df.tail()
```

```python
restricted_tsla_df.to_csv('Bound_TSLA.csv', encoding='utf-8', index=False)
```
