```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/processed_dataset/dataset.csv")
df = df.set_index('Date')
df.head()
```

```python
plt.figure(figsize=(20, 8))
df['Close'].plot()
plt.show()
```

```python
df['PriceDiff'] = df['Close'].shift(-1) - df['Close'] # Calculate price difference between closing prices by day
df['MA10'] = df['Close'].rolling(10).mean() # Calculate moving average based on a 10 day window
df['MA50'] = df['Close'].rolling(50).mean() # Calculate moving average based on a 50 day window
df['LogReturn'] = np.log(df['Close']).shift(-1) - np.log(df['Close']) # Calculate log return
# df = df.dropna()
```
